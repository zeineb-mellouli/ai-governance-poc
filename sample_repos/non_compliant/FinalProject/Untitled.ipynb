```python
import pandas as pd

customers = pd.read_csv('data/customers.csv')
print('Customer data:')
print(customers)
```

```python
top = customers.nlargest(3, 'salary')[['name', 'salary', 'email']]
print('Top earners by salary:')
print(top)
print('\nAverage salary:', customers['salary'].mean())
```
